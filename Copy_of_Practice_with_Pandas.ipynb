Practice Question - The Olympics take place every four years. why are there missing years?

```python
#Find out the missing years
unique_years = oo.Year.unique()
print(unique_years)
```

```python
i=1
missing_years=[]
for i in range(1,len(unique_years)-1):
  curr_year = unique_years[i]
  prev_year = unique_years[i-1]
  while prev_year + 4 < curr_year:
    prev_year +=4
    missing_years.append(int(prev_year))
i +=1
print(missing_years)
```
